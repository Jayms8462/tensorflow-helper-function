# What we've created so far
-Scaler: a single number
-Vector: a number with direction (e.g wind speed and sirection)
-Matrix: a 2-dimensional array of numbers
-Tensor: a n-dimensional array of numbers (When n can be any number, a 0-dimentional tensor is a scalar, a 1-dimentional is a vector)

### Creating Random Tensors
Random tensors are tensors of some abitrary size which contain random numbers.

```python
# Create two random (But the same) tensors
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

# Are they equal? No, ok i was wrong
random_1, random_2, random_1 == random_2
```

# Shuffle the order of elements in a tensor

```python
# Shuffle a tensor (valuable for when you want to shuffle your data so the inherent is not biased)
not_shuffled = tf.constant([[10, 7],
                            [3, 4],
                            [2, 5]])
# Shuffle the above tensor
tf.random.shuffle(not_shuffled)
```

# Getting Information from Tensors
* Shape tensor.shape
* Rank tensor.ndim
* Axis or Dimension tensor[num]
* Size tf.size(tensor)

```python
# Create a rank 4 tensor
rank_4_tensor = tf.zeros([2, 3, 4, 5])
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of out tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in out tensor:", tf.size(rank_4_tensor).numpy())
```

### Indexing Tensors
Tensors can be indexed liek Python Lists.

```python
# Get the first 2 elements of each dimension
rank_4_tensor[:2, :2, :2, :2]
```

```python
# Get the first element from each dimension from each index except for the final one
rank_4_tensor[:1, :1, :1, :]
```

```python
# Create a rank 2 tensor (2 dimensions)
rank_2_tensor = tf.constant([[10, 7],
                             [3, 4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
rank_2_tensor[:, -1]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[..., tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1)
```

```python
# Create a tensor of (3, 2) 
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])
X, Y
```

```python
tf.reshape(Y, shape=(2, 3))
```

```python
Y
```

```python
X @ tf.reshape(Y, shape=(2, 3))
```

```python
tf.reshape(X, shape=(2, 3)) @ Y
```

```python
X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
tf.transpose(X) @ Y
```

```python
# Perform matrix multi between X and Y (Transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform Matrix multi between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```
